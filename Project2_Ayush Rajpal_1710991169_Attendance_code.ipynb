```python
import pandas as pd
import numpy as np

attendance = pd.read_csv(r'C:\Users\Ayush Rajpal\Downloads\Report.csv', header = 5)

attendance.head()
```

```python
attendance.columns
```

```python
attendance.rename(columns={'Unnamed: 0' : 'Serial Number', 'Unnamed: 1' : 'Roll_Num', 'Unnamed: 2' : 'Uni_ID', 
                           'Unnamed: 3' : 'Student Name', 'Unnamed: 31' : 'Total_%age' }, inplace = True)
attendance.columns
```

```python
# There is a space in front of some column names, which is written as '\xa0'
col = attendance.columns
col = list(col)
print(col)
```

```python
# To remove those spaces we will use replace function
attendance.columns = attendance.columns.str.replace('\xa0', '')
attendance.columns
```

```python
attendance.head()
```

```python
attendance.shape
```

```python
attendance.dropna(inplace = True)
attendance.shape
```

```python
attendance['Uni_ID'] = attendance['Uni_ID'].astype(str)
```

```python
# By looking at the values in the raw data, we can see the data also has those values or spaces in front of them
l2 = []
for i in range(20):
    l2.append(attendance['AML3201'][i])
print(l2)
```

```python
# We will again use the replace function to remove those characters
for i in attendance.columns:
    attendance[i] = attendance[i].str.replace('\xa0', '')
```

```python
l3 = []
for i in range(20):
    l3.append(attendance['AML3201'][i])
print(l3)
```

```python
for i in attendance.columns:
    attendance[i] = attendance[i].str.replace('---', '-1')
```

```python
attendance.head()
```

```python
attendance.dtypes
```

```python
for i in attendance.columns:
    if i == "Serial Number" or i == "Roll_Num" or i == "Uni_ID" or i == "Student Name":
        continue
    else:
        attendance[i] = attendance[i].astype(float)
```

```python
attendance["Uni_ID"] = attendance["Uni_ID"].astype(float)
attendance["Uni_ID"] = attendance["Uni_ID"].astype(int)
```

```python
for i in range(0,2):
    attendance[attendance.columns[i]] = attendance[attendance.columns[i]].astype(int)
attendance["Student Name"] = attendance["Student Name"].astype(str)
```

```python
print(attendance.dtypes)
print(attendance.shape)
```

```python
attendance.head()
```

```python
attendance.shape[1]
```

```python
df = [["Roll_Num", "Uni_ID", "Student Name", "Subject", "Percentage"]]
for i in range(attendance.shape[0]):
    for j in range(4, 31):
        if attendance.iloc[i].values[j] < 75 and attendance.iloc[i].values[j] != -1:
            df.append([attendance.iloc[i].values[1], attendance.iloc[i].values[2], attendance.iloc[i].values[3],attendance.columns[j], attendance.iloc[i].values[j]])
            break

Make_up_list_Combined = pd.DataFrame(df[1:], columns = df[0])
Make_up_list_Combined.head()
```

```python
df_Theory = [["Roll_Num", "Uni_ID", "Student Name", "Subject", "Percentage"]]
for i in range(attendance.shape[0]):
    for j in range(4, 27):
        if attendance.iloc[i].values[j] < 75 and attendance.iloc[i].values[j] != -1:
            df_Theory.append([attendance.iloc[i].values[1], attendance.iloc[i].values[2], attendance.iloc[i].values[3],attendance.columns[j], attendance.iloc[i].values[j]])
            break
            
Make_up_list_Theory = pd.DataFrame(df_Theory[1:], columns = df_Theory[0])
Make_up_list_Theory.head()
```

```python
Make_up_list_Theory.to_csv('Make_up_list_Theory.csv')
```

```python
df_Practical = [["Roll_Num", "Uni_ID", "Student Name", "Subject", "Percentage"]]
for i in range(attendance.shape[0]):
    for j in range(27, 31):
        if attendance.iloc[i].values[j] < 75 and attendance.iloc[i].values[j] != -1:
            df_Practical.append([attendance.iloc[i].values[1], attendance.iloc[i].values[2], attendance.iloc[i].values[3],attendance.columns[j], attendance.iloc[i].values[j]])
            break
            
Make_up_list_Practical = pd.DataFrame(df_Practical[1:], columns = df_Practical[0])
Make_up_list_Practical.head()            
```

```python
Make_up_list_Practical.to_csv('Make_up_list_Practical.csv')
```
